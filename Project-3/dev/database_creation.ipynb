```python
import pandas as pd
import os
import csv

from sqlalchemy import create_engine
```

```python
df = pd.read_csv('./attacks.csv', encoding="windows-1252")
```

```python
df.head()
```

```python
# Drop the desired columns
columns_to_drop = ['Location', 'Investigator or Source', 'pdf', 'href formula', 'href', 'Case Number.1', 'Case Number.2', 'original order']
df = df.drop(columns_to_drop, axis=1)

# Save the modified DataFrame back to a CSV file
df.to_csv('cleaned-attacks.csv', index=False)
```

```python
df.head()
```

```python
columns_to_drop = ['Species ','Unnamed: 22', 'Unnamed: 23' ]
df = df.drop(columns_to_drop, axis=1)
```

```python
df.head(50)
```

```python
df['Case Number'].count()
```

```python
#drop rows that dont have data
dropped_missing_data = df.dropna(subset=['Case Number','Date', 'Year', 'Type', 'Country', 'Area' , 'Activity', 'Name', 'Age', 'Injury','Time'])
```

```python
# Delete rows with NaN values in any column
df = df.dropna(how='any')
```

```python
df['Case Number'].count()
```

```python
df.head(50)
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')
df = df.dropna(subset=['Date'])
```

```python
df.head(10)
```

```python
df = df.drop('Time', axis=1)
```

```python
df.head()
```

```python
df.columns
```

```python
df['Year'] = df['Year'].astype(int)
```

```python
df.head()
```

```python
# Convert the 'Age' column to numeric, coerce non-numeric values to NaN
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Remove rows with non-numeric values in 'Age' column
df = df.dropna(subset=['Age'])
```

```python
df.head(80)
```

```python
df['Age'] = df['Age'].astype(int)
```

```python
df.head()
```

```python
columns = {'Case Number':str, 'Date':str, 'Year':int, 'Type':str, 'Country':str, 'Area':str, 'Activity':str,
       'Name':str, 'Sex ':str, 'Age':int, 'Injury':str, 'Fatal (Y/N)':str}

df = df.astype(columns)
```

```python
df.columns
```

```python
df.columns = ['Case_Number', 'Date', 'Year', 'Type', 'Country', 'Area', 'Activity',
       'Name', 'Sex', 'Age', 'Injury', 'Fatal']
```

```python
df.columns
```

```python
df['Sex'] = df['Sex'].str.strip()
```

```python
df = df[df['Sex'] != 'lli']
```

```python
value_counts = df['Sex'].value_counts()
value_counts
```

```python
# Save the modified DataFrame back to a CSV file
df.to_csv('cleaned-attacks.csv', index=False)
```

```python
# Create database connection
connection_string= 'postgresql+psycopg2://postgres:1357@localhost:5432/Attacks_db'
engine = create_engine(connection_string)
```
